Data cleaning:

```python
data.duplicated().sum()                     #5 duplicates
```

```python
data=data.drop_duplicates()
```

```python
data.duplicated().sum()
```

```python
data=data.sort_values(by="patient_id",ascending=True)         #arange rows serially
```

```python
data=data.reset_index(drop=True)
```

```python
data.isna().sum()
```

```python
data['age']=data['age'].fillna(data['age'].median())                  #fill null values with median
```

```python
cols_to_fill=data[['creatinine_phosphokinase','ejection_fraction','platelets','serum_creatinine','serum_sodium']]
for col in cols_to_fill:
    data[col]=data[col].fillna(data[col].median())
```

```python
data.drop(columns='patient_notes',inplace=True)                    #huge null vallues so drop column
```

```python
data.isna().sum()
```

```python
data['age_group']=pd.cut(data['age'],bins=[0,50,60,70,120],labels=['<50','50-60','60-70','70+'])                
```

```python
data.head()
```

```python
data['sex_label']=data['sex'].str.capitalize().str.strip()
```

```python
data['sex']=data['sex_label'].map({'Male':1,'Female':0})
```

```python
data.head()
```

```python
data['ejection_risk']=pd.cut(data['ejection_fraction'],bins=[0,30,45,100],labels=['High Risk','Moderate Risk','Normal'])
```

```python
data['creatinine_risk']=pd.cut(data['serum_creatinine'],bins=[0,1.0,1.5,20],labels=['Normal','Borderline','Elevated'])
```

```python
data.head()
```

```python
import pandas as pd
import matplotlib.pyplot as plt
import seaborn as sns
```

```python
df=pd.read_csv(r"C:\Users\Hp\Data Analyst Internship\Cleaned Heart Disease Data.csv")
```

```python
df.head()
```

```python
df.shape
```

```python
df.describe(include="all")
```

```python
df.duplicated().sum()
```

```python
fig,axs=plt.subplots(2,3,figsize=(10,5))

axs[0,0].text(0.5,0.5,len(df),fontsize=20,ha='center')
axs[0,0].set_title("Total patients:",fontweight='bold',fontsize=15)
axs[0,0].axis('off')

axs[0,1].text(0.5,0.5,df['sex'].sum(),fontsize=20,ha='center')
axs[0,1].set_title("Male patients:",fontweight='bold',fontsize=15)
axs[0,1].axis('off')

axs[0,2].text(0.5,0.5,(df['sex']==0).sum(),fontsize=20,ha='center')
axs[0,2].set_title("Female patients:",fontweight='bold',fontsize=15)
axs[0,2].axis('off')

axs[1,0].text(0.5,0.5,len(df)-df['Death_Event'].sum(),fontsize=20,ha='center')
axs[1,0].set_title("Survived:",fontweight='bold',fontsize=15)
axs[1,0].axis('off')

axs[1,1].text(0.5,0.5,df['Death_Event'].sum(),fontsize=20,ha='center')
axs[1,1].set_title("Deaths:",fontweight='bold',fontsize=15)
axs[1,1].axis('off')

mortality_rate=df['Death_Event'].mean() * 100
axs[1,2].text(0.5,0.5,f"{mortality_rate:.2f}%",fontsize=20,ha='center')
axs[1,2].set_title("Overall mortality rate:",fontweight='bold',fontsize=15)
axs[1,2].axis('off')

plt.tight_layout()
plt.show()
```

```python
plt.figure(figsize=(8,5))

outcome=df['Death_Event'].value_counts()
plt.pie(outcome,labels=['Survived','Dead'],autopct="%1.1f%%",startangle=90,colors=["#16A34A","#DC2626"])
plt.title("Patient Outcome Distribution")
plt.show()
```

Interpretation:32% patients died among 299 patients.This shows 1 in 3 patients did not survive.This mortality date is highly critical we should take immediate action over this.

```python
import seaborn as sns
plt.figure(figsize=(12,7))
sns.histplot(data=df[df['Death_Event']==0],x='age',alpha=0.8,label="Survived")
sns.histplot(data=df[df['Death_Event']==1],x='age',alpha=0.7,label="Dead")

plt.title("Age distribution by outcome")
plt.xlabel("Age")
plt.ylabel("Patient count")
plt.legend()
plt.tight_layout
plt.show()
```

Interpretation-Deaths are concentrated in 60-80 age range.Patients below age 55 shows significantly good survival rate.This shows older patients should be given close monitoring.

```python
sns.boxplot(data=df,x='Death_Event',y='ejection_fraction')
plt.title("Ejection Fraction by Outcome")
plt.xticks(ticks=[0,1],labels=['Survived','Died'])
plt.show()
```

```python
ef_high=df[df['ejection_fraction'] < 30]['Death_Event'].mean() * 100
ef_norm=df[df['ejection_fraction'] >= 45]['Death_Event'].mean() * 100

print(f"Patients with EF < 30% = {ef_high:.1f}%")
print(f"Patients with EF >= 45% = {ef_norm:.1f}%")
```

Interpretation-Patients who died had lower median ejection fraction compared to survivors.Shows ejection fraction is a strong predictor of mortality. Patients with EF below 30% face 64.9% mortality vs 25.0% for patients with EF above 45%.

```python
sns.boxplot(data=df,x='Death_Event',y='serum_creatinine')
plt.title("Serum Creatinine by Outcome")
plt.xticks(ticks=[0,1],labels=['Survived','Died'])
plt.show()
```

```python
cr_high=df[df['serum_creatinine'] > 1.5]['Death_Event'].mean() * 100
cr_norm=df[df['serum_creatinine'] <= 1.0]['Death_Event'].mean() * 100

print(f"Patients with creatinine > 1.5 = {cr_high:.1f}%")
print(f"Patients with creatinine <= 1.0 = {cr_norm:.1f}%")
```

Interpretation-Patients who died show significantly higher serum creatinine levels with multiple extreme outliers in the died group.High creatinine indicates kidney disfunction which directly worsens heart failure outcomes.Patients with creatinine above 1.5 face 65.2% mortality vs 17.2% for those at or below 1.0

```python
gender=df.groupby('sex_label')['Death_Event'].agg(['sum', 'count'])
gender['survival_rate']=(gender['count'] - gender['sum']) / gender['count'] * 100
gender['mortality_rate']=gender['sum']/gender['count'] * 100
print(gender)
```

```python
plt.figure(figsize=(10,8))
gender[['survival_rate','mortality_rate']].plot(kind='bar',color=['steelblue', 'crimson'])

plt.ylabel("Rate (%)")
plt.title("Mortality vs Survival Rate by Gender")
plt.legend(['Survival %','Mortality %'],loc='upper right')
plt.tight_layout()
plt.grid(linestyle='--',linewidth=1)
plt.show()
```

Interpretation-Females show a slightly higher mortality rate than males.Both genders face similar mortality risk,suggesting gender alone is not a primary predictor.Other factors like ejection fraction and creatinine are far stronger predictors than gender."

```python
plt.figure(figsize=(7,7))
num_cols=['age','creatinine_phosphokinase','ejection_fraction','platelets','serum_creatinine','serum_sodium','time','Death_Event']

corr=df[num_cols].corr()
sns.heatmap(corr,annot=True,cmap='coolwarm',fmt='.2f')
```

Interpretation-The correlation heatmap reveals that serum creatinine and ejection fraction are the two strongest correlators with death event among all clinical features.Age shows positive correlation confirming older patients are at higher risk.Time shows negative correlation showing patients who died early naturally had shorter follow-up periods.

```python
age_mort=df.groupby('age_group')['Death_Event'].mean() * 100
age_mort=age_mort.reindex(['<50', '50-60', '60-70', '70+'])

plt.bar(age_mort.index, age_mort.values,color=['steelblue','orange','crimson','purple'])
plt.title("Mortality Rate by Age Group")
plt.ylabel("Mortality Rate (%)")
plt.xlabel("Age Group")
plt.tight_layout()
plt.show()
```

```python
age70_mort=df[df['age'] >= 70]['Death_Event'].mean() * 100
age50_mort=df[df['age'] < 50]['Death_Event'].mean() * 100

print(f"Patients aged 70+ = {age70_mort:.1f}%")
print(f"Patients aged < 50 = {age50_mort:.1f}%")
```

Interpretation-Mortality rate increases consistently with age.Patients having age > 70 should be given priority becauze their death rate is higher.
Patients aged 70+ face 49.7% mortality — nearly double the rate of patients below 50 (22.9%).

```python
ej_mort=df.groupby('ejection_risk')['Death_Event'].mean() * 100

plt.bar(ej_mort.index, ej_mort.values,color=['crimson','orange','green'])
plt.title("Mortality Rate by Ejection Fraction Risk")
plt.ylabel("Mortality Rate (%)")
plt.xlabel("Risk Category")
plt.grid(axis='y',linestyle='--',linewidth=1)
plt.tight_layout()
plt.show()
```

Interpretation-Patients in the High Risk ejection fraction category (EF below 30%)face 64.4% mortality — nearly 3 times higher than patients in the Normal category (23.8%).Any patient admitted with EF below 30% must be treated as critical priority.

```python
binary_cols=['anaemia','diabetes','high_blood_pressure','smoking']
binary_labels=['Anaemia','Diabetes','High BP','Smoking']

mort_with=[df[df[c]==1]['Death_Event'].mean()*100 for c in binary_cols]
mort_without=[df[df[c]==0]['Death_Event'].mean()*100 for c in binary_cols]

comorbidity = pd.DataFrame({'Has Condition':mort_with,'No Condition':mort_without},index=binary_labels)
comorbidity.plot(kind='bar',color=['crimson','steelblue'],figsize=(10,5))

death_rate=df['Death_Event'].mean() * 100
plt.axhline(death_rate,color='gray',linestyle='--',linewidth=2,label=f'Overall avg:{death_rate:.1f}%')
plt.ylabel("Mortality Rate (%)")
plt.title("Mortality:With vs Without Comorbidities")
plt.xticks(rotation=0)
plt.legend()
plt.tight_layout()
plt.show()
```

Interpretation-Among all four comorbidities,high BP shows the highest impact on mortality,patients with high BP die at higher rate than those without.Anaemia ,Diabetes and smoking all show moderate increases in mortality compared to patients without those conditions.Suggests comorbidities are contributing factors but not primary predictors.

```python
df['risk_score']=((df['ejection_fraction'] < 30).astype(int) * 2 +
                (df['serum_creatinine'] > 1.5).astype(int) * 2 +
                (df['serum_sodium'] < 130).astype(int) +
                (df['high_blood_pressure'] == 1).astype(int) +
                (df['anaemia'] == 1).astype(int) +
                (df['diabetes'] == 1).astype(int) +
                (df['age'] > 65).astype(int)
)

df['outcome_label']=df['Death_Event'].map({0:'Survived',1:'Died'})
```

```python
plt.figure(figsize=(8,5))
sns.histplot(data=df,x='risk_score',hue='outcome_label',multiple='dodge',palette={'Survived':'steelblue','Died':'crimson'})

plt.title("Risk Score Distribution by Outcome")
plt.xlabel("Composite Risk Score")
plt.ylabel("Patient Count")
plt.tight_layout()
plt.show()
```

Interpretation-At low risk scores (0–2),survived patients are more.As the score increases beyond 4,the died group begins to dominate.This shows that the composite risk score successfully separates high risk patients from low risk ones,patients with higher scores are lying in the died group.

```python
df['risk_band']=pd.cut(df['risk_score'],bins=[-1, 1, 3, 5, 10],labels=['Low (0-1)','Medium (2-3)','High (4-5)','Critical (6+)'])

band_mort=df.groupby('risk_band',observed=True)['Death_Event'].mean() * 100

plt.figure(figsize=(8,5))
plt.bar(band_mort.index, band_mort.values,color=['green','orange','crimson','darkred'])
plt.title("Mortality Rate by Patient Risk Band")
plt.ylabel("Mortality Rate (%)")
plt.xlabel("Risk Band")
plt.tight_layout()
plt.show()
```

Interpretation-Mortality rate rises sharply across bands,from low to critical.Patients in the Critical band (score 6+) face 77.8% mortality.These risk scores can be used to immediately identify which patients need ICU-level care and which can be managed through standard protocols."
